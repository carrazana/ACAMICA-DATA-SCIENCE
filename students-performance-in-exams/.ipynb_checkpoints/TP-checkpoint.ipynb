```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.3f' % x)
path_dataset = 'StudentsPerformance.csv'
df = pd.read_csv(path_dataset)
```

```python
df.head()
```

```python
df['test_mean']=((df['math score']+df['reading score']+df['writing score'])/3)
```

```python
pd.get_dummies(df, columns = ["gender","lunch","test preparation course"])
```

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

df.gender.unique()
```

```python
le = LabelEncoder()
df['race/ethnicity'] = le.fit_transform(df['race/ethnicity'])

df['race/ethnicity'].unique()
```

```python
le = LabelEncoder()
df['parental level of education'] = le.fit_transform(df['parental level of education'])

df['parental level of education'].unique()
```

```python
le = LabelEncoder()
df['lunch'] = le.fit_transform(df['lunch'])

df['lunch'].unique()
```

```python
le = LabelEncoder()
df['test preparation course'] = le.fit_transform(df['test preparation course'])

df['test preparation course'].unique()
```

```python
df.head()
```

```python
df.corr()
```

```python
X = df.drop(['parental level of education'], axis=1)
y = df['parental level of education']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

knn = KNeighborsClassifier()
knn.fit(X_train, y_train) 
y_predict = knn.predict(X_test)
#rmse = root_mean_squared_error(y_test, y_predict)
accuracy_score(y_test, y_predict)
```
